```python
# Set country as index
data1.set_index('country', inplace=True)
```

```python
data_Lesotho = data1.loc['Lesotho']
data_Lesotho
```

```python
# Top 5 countries
data_Mozambique   = data1.loc['Mozambique']
data_Gabon        = data1.loc['Gabon']
data_GuineaBissau = data1.loc['Guinea-Bissau']
data_SouthAfrica  = data1.loc['South Africa']

# Plus INdia
data_India = data1.loc['India']
data_Bangladesh = data1.loc['Bangladesh']
```

```python
x_values = [i for i in range(year_min, year_max+1)]

y_values_L  = data_Lesotho.e_mort_100k
y_values_M  = data_Mozambique.e_mort_100k
y_values_G  = data_Gabon.e_mort_100k
y_values_GB = data_GuineaBissau.e_mort_100k
y_values_SA = data_SouthAfrica.e_mort_100k

y_values_I  = data_India.e_mort_100k
y_values_B  = data_Bangladesh.e_mort_100k
```

```python
plt.ylim(0, 600)
plt.xlim(year_min, year_max)

plt.plot(x_values, y_values_L,  color = 'navy',       label = 'Lesotho',      marker = "x")
plt.plot(x_values, y_values_M,  color = 'darkred',    label = 'Mozambique',   marker = "1")
plt.plot(x_values, y_values_SA, color = 'darkorange', label = 'South Africa', marker = "2")
plt.plot(x_values, y_values_G,  color = 'steelblue',  label = 'Gabon',        marker = "3")
plt.plot(x_values, y_values_GB, color = 'darkgreen',  label = 'GuineaBissau', marker = "4")
#plt.plot(x_values, y_values_I, color = 'black', label = 'India', marker = "+")

plt.xlabel('Year', fontsize=15)
plt.ylabel('Estimated mortality', fontsize=15)
plt.title('Estimated mortality of TB cases (all forms) per 100 000 population: Top 5 Countries in 2017', fontsize=15)
#plt.legend(title = 'Top 5 Countries in 2017', fontsize=15)
plt.legend(fontsize=15)

plt.savefig("TBmortalityTop5Countries")
plt.show()
```
